```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load your dataset
# Example: dataset = pd.read_csv('your_file.csv')

# Define categorical columns to encode
categorical_columns = [
    'workclass',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'gender',
    'native-country'
]

# Dictionary to store LabelEncoders for each column
encoders = {}

# Apply Label Encoding
for col in categorical_columns:
    le = LabelEncoder()

    # Fit and transform the column
    dataset[col] = le.fit_transform(dataset[col].astype(str))  # Ensure all entries are strings

    # Save the encoder for later use (optional)
    encoders[col] = le

# View sample of transformed data
print(dataset.head())

# Optionally: print label mappings for each column
for col in categorical_columns:
    print(f"\nLabel Mapping for '{col}':")
    labels = encoders[col].classes_
    for idx, label in enumerate(labels):
        print(f"{idx}: {label}")
```
